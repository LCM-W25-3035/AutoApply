```python
'''
create the code for extracting the data from octoparse
'''

# This code is used when there is paid version of Octoparse
import requests
import pandas as pd

# Octoparse API Key 
API_KEY = ""

# Task ID 
TASK_ID = ""

# Octoparse API URL to fetch data
API_URL = f"https://dataapi.octoparse.com/api/alldata/{TASK_ID}?token={API_KEY}"

# Fetch data from Octoparse API
response = requests.get(API_URL)

# Check if request is successful
if response.status_code == 200:
    data = response.json()
    
    # Extract records
    job_data = data.get("data", [])
    
    if job_data:
        # Convert to DataFrame
        df = pd.DataFrame(job_data)
        
        # Save to CSV
        csv_filename = "octoparse_scraped_jobs.csv"
        df.to_csv(csv_filename, index=False)

        print(f"✅ Job data extracted and saved as {csv_filename}")
    else:
        print("❌ No job data found in Octoparse task.")
else:
    print(f"❌ Failed to fetch data. Error Code: {response.status_code}")
```

```python
from jobspy import scrape_jobs
import pandas as pd

# Scrape Data Engineer job postings from Indeed (Ontario) - Last 14 Days
df = scrape_jobs(
    site_name="indeed",  # Use Indeed instead of LinkedIn
    search_term="Data Engineer",
    location="Ontario",
    results_wanted=50,
    days_old=14,
    scrape_description=True  # Ensure descriptions are extracted
)

# Check if 'description' column exists
if "description" in df.columns:
    df["job_description"] = df["description"]
else:
    df["job_description"] = "No job description available."

# Save the extracted data
csv_filename = "indeed_data_engineer_jobs.csv"
df.to_csv(csv_filename, index=False)

# Display confirmation
print(f"✅ Job postings saved as {csv_filename} (With extracted job descriptions)")

# Show first few rows
df[['title', 'job_description']].head()
```

```python
from jobspy import scrape_jobs
import pandas as pd

# Scrape Data Engineer job postings from LinkedIn (Ontario) - Last 14 Days
df = scrape_jobs(
    site_name="linkedin",
    search_term="Data Engineer",
    location="Ontario",
    results_wanted=50,
    days_old=14,
    scrape_description=False  # No need to scrape descriptions since jobspy doesn't extract them
)

# Extract relevant columns (Job Title, Company, Location, Job URL)
df_links = df[['title', 'company', 'location', 'job_url_direct', 'company_description']]

# If 'company_description' contains job details, use it
df_links['job_description'] = df_links['company_description'].fillna("Job description not available.")

# Save job URLs for manual review or additional scraping
csv_filename = "linkedin_job_links.csv"
df_links.to_csv(csv_filename, index=False)

print(f"✅ Job posting links saved as {csv_filename}. Use these links to access full job descriptions.")
df_links.head()
```
